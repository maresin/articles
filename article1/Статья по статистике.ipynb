# Базовая статистика на одном примере

### Цель статьи
Объединить разрозненные теоретические знания в рамках исследования одной задачи. 

### Задача
> "У вас сеть станций проката «Самокат подвезёт»: 20 точек в центре города, на каждой — максимум 50 электросамокатов. Вы хотите проверить гипотезу, что в прошлом месяце на любой станции днём было доступно в среднем 30 самокатов. Именно о таком количестве говорили урбанисты КБ «Белка» в исследовании мобильности горожан. Если самокатов на станции меньше, то пользователи думают, что не смогут арендовать самокат, когда он понадобится. А когда их больше, горожанам кажется, что оставить самокат на парковке не получится — не будет свободных мест. Каждая станция ежечасно отсылала на сервер текущее количество самокатов. Вы загрузили значения с 13:00 до 16:00 за последние 30 дней. Проверьте гипотезу по этой выборке".

### Этапы решения
* Формирование гипотез
* Построение гистограммы выборки и расчет ее параметров
* Переход от параметров выборки к соответствующим оценкам параметров генеральной совокупности
* Центральная предельная теорема
* Параметры нормального распределения выборочных средних
* Правило трех сигмы и пороги значимости
* Оценка отклонения выборочного среднего и выбор гипотез
* Доверительный интервал
* Распределение Стьюдента
* Сравнение двух выборок

### Решение

#### Формирование гипотез
Суть задачи сводится к доказательству статистической значимости наблюдаемых результатов. Проще говоря, к доказательству того, что данные выборки отличаются от данных генеральной совокупности. При этом, с практической точки зрения, нам бы хотелось доказать утверждение: _"Если самокатов на станции меньше, то пользователи думают, что не смогут арендовать самокат, когда он понадобится. А когда их больше, горожанам кажется, что оставить самокат на парковке не получится — не будет свободных мест"_. Доказать его мы, однако, не можем. Но, мы можем выдвинуть две гипотезы - основную и альтернативную (желательную), и постараться опровергнуть основную (или нулевую гипотезу). Нулевая гипотеза всегда предполагает, что две сравниваемые совокупности не различаются между собой. Альтернативная гипотеза всегда противопоставляется нулевой. Ни о каком доказательстве не может идти речи, если нулевая и альтернативная гипотеза не покрывают все пространство возможных вариантов. Этот процесс заканчивается только тогда, когда остается гипотеза, которой не находится альтернативы.

__Нулевая гипотеза $H_0$:__ Среднее генеральной совокупности равно 30 или $E(\overline X)=30$.  
__Альтернативная гипотеза $H_1$:__ Среднее генеральной совокупности не равно 30 или $E(\overline X)\neq30$.

```python
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
scooters = np.array([15, 31, 10, 21, 21, 32, 30, 25, 21,
28, 25, 32, 38, 18, 33, 24, 26, 40, 24, 37, 20, 36, 28, 38,
24, 35, 33, 21, 29, 26, 13, 25, 34, 38, 23, 37, 31, 28, 32,
24, 25, 13, 38, 34, 48, 19, 20, 22, 38, 28, 31, 18, 21, 24,
31, 21, 28, 29, 33, 40, 26, 33, 33,  6, 27, 24, 17, 28,  7,
33, 25, 25, 29, 19, 30, 29, 22, 15, 28, 36, 25, 36, 25, 29,
33, 19, 32, 32, 28, 26, 18, 48, 15, 27, 27, 27,  0, 28, 39,
27, 25, 39, 28, 22, 33, 30, 35, 19, 20, 18, 31, 44, 20, 18,
17, 28, 17, 44, 40, 33,])
```

#### Построение гистограммы выборки и расчет ее параметров
Для начала взглянем на наши данные и заодно вспомним о там, что стоит различать два вида гистограмм - частотную и вероятностную. По графикам можно примерно определить вероятность попадания значений в тот или иной интервал. Например, для интервала значений от 25 до 30 вероятность составит около 0.3.

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
ax[0].hist(scooters, color='0.5')
ax[0].set_title('Частотная гистограмма выборки')
ax[0].set_ylabel('Количество')
ax[0].set_xlabel('x')
ax[0].text(3, 25, '$p=H_{ст.} /\ n$', fontsize=14)
ax[0].fill_between([24, 29], 35, 0, color='0.7')

ax[1].hist(scooters, density=True, color='0.5')
ax[1].set_title('Гистограмма плотности вероятности выборки')
ax[1].set_ylabel('Плотность вероятности')
ax[1].set_xlabel('x')
ax[1].text(3, 0.04, '$p=H_{ст.} *\ W_{ст.}$', fontsize=14)
ax[1].fill_between([24, 29], 0.061, 0, color='0.7')
plt.show;
```

Если говорить о непрерывном распределении, то тут вероятность придется рассчитывать как определенный интеграл. Хотя и здесь какую-то примерную оценку можно дать.

```python
plt.figure(figsize=(6, 5))
x = np.linspace(6, 54, 100)
x_tr = np.linspace(23, 30, 2)
y_tr = x_tr/630
plt.plot(x, st.norm.pdf(x, 30, 8.4), color='0.5')
plt.fill_between(x=[24, 29], y1=y_tr, y2=0, color='0.6')
plt.title('Непрерывное распределение')
plt.xlabel('$X$')
plt.ylabel('Плотность вероятности')
plt.text(30, 0.03, r'$p \approx \frac {(H_{ст.1} + H_{ст.2})*W_{ст.}} {2}$', fontsize=14)
plt.show;
```

#### Переход от параметров выборки к соответствующим оценкам параметров генеральной совокупности
Рассчитаем параметры выборки - выборочное среднее и дисперсию. Расчет выборочного среднего происходит обычным образом. Дисперсия обычно считается как:
$$\sigma^2 = \frac {\displaystyle\sum_{i=1}^{n} {x_i - \mu}} {n}$$
Однако дисперсия выборки мало что говорит о генеральной совокупности. В формулу вносится поправка Бесселя, что несколько увеличивает дисперсию:
$$s^2 = \frac {\displaystyle\sum_{i=1}^{n} {x_i - \overline x}} {n-1}$$
Итак, в результате расчетов среднего и дисперсии выборки мы получили несмещенные оценки среднего и дисперсии уже генеральной совокупности. Заметим, что речь идет не о точных значениях, а лишь об оценках, мататожиданиях случайных величин. После извлечения корня из дисперсии можно получить выборочное стандартное отклонение.

```python
scooters_m = 30
print('Предполагаемое среднее генеральной совокупности:', scooters_m)
scooters_size = len(scooters)
print('Размер выборки:', scooters_size)
scooters_mean = scooters.mean()
print('Выборочное среднее (несмещенная оценка среднего генеральной совокупности):', scooters_mean)
print('Стандартное отклонение выборки:', scooters.std(ddof=0))
scooters_std = scooters.std(ddof=1)
print('Несмещенная оценка стандартного отклонения генеральной совокупности:', scooters.std(ddof=1))
```

#### Центральная предельная теорема
После получения оценочных данных для среднего и стандартного отклонения генеральной совокупности вспоминается колоколообразное распределение, правило трех сигм и возникает большой соблазн тут же броситься считать вероятности. Однако, что дает нам основание считать, что значения в генеральной совокупности распределены нормально? А ведь многие пытаются провернуть этот фокус еще раньше - прямо на гистограмме выборки. А там, уж точно, ни о каком нормальном распределении речь не идет.  
Здесь можно столкнуться с еще с одной ошибкой - использованием в качестве среднего значения генеральной совокупности его несмещенной оценки. На самом деле в этом качестве должно выступать предполагаемое среднее.  
И не стоит забывать, что для определения вероятности нам нужен график плотности вероятности, а не привычная частотная гистограмма. Не подойдет даже гистограмма плотности вероятности из-за того, что мы не знаем точных значений в генеральной совокупности. Все, чем располагаем - это предполагаемая функция плотности вероятности.  
И так мы плавно переходим к ЦПТ - Центральной предельной теореме. Суть которой в том, что если многократно взять из генеральной совокупности выборки одного размера и посчитать их средние, то они образуют распределение, нормально расположенное относительно среднего генеральной совокупности (выверты терминологии: такое распределение - это выборочное среднее, только не число, а случайная величина).

Собственно функция, описывающая нормвльное распределение:
$$f(x)=\frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$$

```python
scooters_se = scooters_std / (scooters_size ** .5)
print('Стандартная ошибка среднего:', scooters_se)
scooters_z = (scooters_mean - scooters_m) / scooters_se
print('Стандартизированное отклонение среднего (Z-статистика):', scooters_z)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
x0 = np.linspace(scooters_m - 4*scooters_se, scooters_m + 4*scooters_se, 100)
ax[0].plot(x0, st.norm.pdf(x0, scooters_m, scooters_se), color='0.4', linewidth=5, label='$f(x)$')
ax[0].vlines(scooters_m, 0, 0.5, color='0.7', linewidth=3, linestyle = ':', label='$M_{h0}$')
ax[0].vlines(scooters_mean, 0, 0.5, color='0.7', linewidth=3, linestyle = '--', label='$X_{s}$')
ax[0].set_title('Нормальное \nвыборочное распределение')
ax[0].set_xlabel('$X$')
ax[0].set_ylabel('Плотность вероятности')
ax[0].legend()
x1 = np.linspace(- 4, 4, 100)
ax[1].plot(x1, st.norm.pdf(x1, 0, 1), color='0.4', linewidth=5, label='$f(x)$')
ax[1].vlines(0, 0, 0.4, color='0.7', linewidth=3, linestyle = ':', label='$M_{h0}$')
ax[1].vlines(scooters_z, 0, 0.4, color='0.7', linewidth=3, linestyle = '--', label='$Z_{s}$')
ax[1].set_title('Нормальное стандартизированное \nвыборочное распределение')
ax[1].set_xlabel('$Z$')
ax[1].set_ylabel('Плотность вероятности')
ax[1].legend()
plt.show;
```

#### Параметры нормального распределения выборочных средних
Подобное выборочное распределение, как любое нормальное распределение характеризуют два параметра - среднее и стандартное отклонение. Поскольку мы полагаем, что нулевая гипотеза верна, то в качестве среднего выступает предполагаемое среднее значение генеральной совокупности, а именно - 30. В качестве второго параметра выступает стандартная ошибка среднего, которая рассчитывается на основе несмещенной оценки стандартного отклонения генеральной совокупности, полученной ранее. А именно:
$$SE_{\overline X} = \frac{s} {\sqrt n}$$  
На этом этапе уже можно проводить сравнение значения выборочного среднего (иначе несмещенной оценки среднего генеральной совокупности) и предполагаемого среднего генеральной совокупности. Если бы мы делали множество выборок и получили множество выборочных средних, то можно было бы построить распределение. Его обычно подвергают процедуре стандартизации, а точнее Z-стандартизации. Результатом чего является распределение со средним, равным 0 и стандартным отклонением, равным 1. Мы полагаем его нормальным, так как наша выборка достаточно велика и репрезентативна. В нашем случае распределение лишь гипотетическое - мы располагаем всего одним значением выборочного среднего. Все же посчитаем для него Z-статистику (заметим, что статистикой можно назвать любые характеристики, полученные на выборке). Она будет измерятся в сигмах или, точнее, в стандартных ошибках:
$$Z_{\overline X} = \frac{\overline X - M} {SE_{\overline X}}$$

#### Правило трех сигмы и пороги значимости
Итак, мы добрались до графика плотности вероятности и можем посчитать вероятность того, что выборочное среднее отклониться от среднего нашего распределения на величину большую или равную полученной. И равна она площади соответствующего участка под кривой. Чтобы бы ее вычислить нужно решить несобственный интеграл от функции нормального распределения. Вручную это сделать вряд ли удастся. Но даже не зная точного значения вероятности, а располагая лишь некоторыми знаниями о нормальном распределении уже можно будет принять решение о судьбе нулевой гипотезы.
Тут, наконец, в дело вступает правило 3 сигм, которое гласит, что почти все значения нормально распределенной случайной величины лежат в диапазоне $\pm3$ стандартных отклонения ($\pm 3\sigma$ или, в нашем случае, $\pm3 SE_{\overline X}$). В практических задачах мы не можем полагать, что статистической значимостью обладают лишь значения отклонений превышающие эти пределы. Поэтому вводится порог статистической значимости. Его принято измерять в процентах размера выборки, а не в сигмах. Дополним наши знания о нормальном распределении: $95\%$ его значений лежит в пределах $\pm1.96$ сигм, а $99\%$ - в пределах $\pm2.5$ сигм. Эти значения соответствуют обычно принятым порогам значимости $5\%$ и $1\%$ соответственно.

```python
alpha = 0.05
print('Порог статистической значимости:', alpha)
lim_alpha = 1.96
print('Предельное оклонение для порога статистической значимости:', lim_alpha)
```

#### Оценка отклонения выборочного среднего и выбор гипотез
Посмотрим, превышает ли полученное Z-значение величину в 1.96. Очевидно, что $|-3.69| > 1.96$. То есть:
$$|Z_{\overline X}| > Lim_{\alpha}$$
Оно оказалось больше - можем отклонить нулевую гипотезу, так как ее вероятность оказалась ниже $5\%$-го порога значимости.
Такая вероятность носит название `P-value` и показывает вероятность получения наблюдаемых результатов при условии, что нулевая гипотеза верна, или вероятность ошибки в случае отклонения нулевой гипотезы.

#### Доверительный интервал
Проделанный нами путь был необходим, чтобы дать точечную оценку исследуемого значения. Но можем сделать еще и интервальную, когда точное значение нам не известно. Введем понятие доверительного интервала - так называют интервал, который покрывает неизвестный параметр с заданной надёжностью. Будем исходить из того, что наш этот неизвестный параметр (среднее генеральной совокупности) должен попадать в доверительный интервал с $95\%$-ой вероятностью. Что соответствует отклонению в $\pm1.96$ сигм от нашего выборочного среднего. 
$$\overline X - 1.96 \cdot SE_{\overline X}  < M < \overline X + 1.96 \cdot SE_{\overline X}$$  
В нашем случае гипотетическое среднее не попадает в доверительный интервал.

```python
conf_interval = np.round([scooters_mean - 1.96*scooters_se, scooters_mean + 1.96*scooters_se], 2)
print(f'Доверительный интервал: ({conf_interval[0]}, {conf_interval[1]})' )
```

#### Распределение Стьюдента
Пришла пора вспомнить о t-тестах и t-распределении Стьюдента. До сих пор мы полагали, что выборочные средние (из которых мы взяли только одну) формируют нормальное распределение. Однако, при малых объемах выборок распределение видимо отличается от нормального. Отклонения от результирующего среднего будут чаще - распределение будет иметь тяжелые "хвосты", а известные нам закономерности нормального распределения уже не будут соблюдаться. В этом случае нам больше подходит закон t-распределения Стьюдента, которое на больших выборках превращается в нормальное. Сделанные нами ранее выкладки в целом справедливы и для t-распределения. С той разницей, что значения отклонений в сигмах будут больше, и будут зависеть от точного числа наблюдений в выборках, или от числа степеней свободы  - в терминах t-распределении Стьюдента:
$$\mathrm{df} = n - 1$$
Информацию о них можно найти в соответствующих [таблицах значений](https://ru.wikipedia.org/wiki/Распределение_Стьюдента). Применительно к распределению Стьюдента z-значение отклонения выборочного среднего теперь именуется t-критерием.

Сама функция t-распределения имеет еще более громоздкий вид, чем функция нормального:
$${\displaystyle f_{t}(x)={\frac {\Gamma \left({\frac {n}{2}}\right)}{{\sqrt {{(n-1)}\pi }}\,\Gamma \left({\frac {n-1}{2}}\right)}}\,\left(1+{\frac {x^{2}}{n-1}}\right)^{-{\frac {n}{2}}}}$$

```python
fig, ax = plt.subplots(figsize = (6, 5))
x = np.linspace(-4, 4, 100)
ax.plot(x, st.norm.pdf(x, 0, 1), color='0.7', linestyle='--', linewidth=3)
ax.plot(x, st.t.pdf(x, df=3), color='0.5', linewidth=5)
ax.set_title('Форма t-распределения при n = 3')
plt.xlabel('$X$')
plt.ylabel('Плотность вероятности')
plt.show;
```

Собственно сам t-тест для нашей задачи.

```python
results = st.ttest_1samp(scooters, scooters_m)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Сравнение двух выборок
В нашей задаче мы оценивали равенство математического ожидания среднего $E(X)$ некоторому известному значению $M$. Для случая сравнения двух выборок измениться лишь формула расчета t-критерия:
$$\displaystyle t={\frac {{\overline {X}}_{1}-{\overline {X}}_{2}}{\sqrt {{\frac {s_{1}^{2}}{n_{1}}}+{\frac {s_{2}^{2}}{n_{2}}}}}}$$
